```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/lustre/gmeteo/WORK/DATA/C3S-CDS/ERA5_temp/c3s-cds/catalogues/all_catalogues.csv')

# Display the first few rows
df.head()
```

```python
# Get all unique datasets
unique_datasets = df['dataset'].unique()

# Print the unique datasets
print(unique_datasets)
```

```python
grouped_datasets = df.groupby('dataset_type')['dataset'].unique()
print(grouped_datasets)
```

```python
# Filter the dataset for 'raw' product type
raw_data = df[df['product_type'] == 'raw']

raw_data.head()
```

```python
# Retrieve the data paths for all raw variables
raw_paths = raw_data[['variable', 'data_path']]
raw_paths.head()
```

```python
# Filter by a specific variable, e.g., 't2m', and get the paths
t2m_raw_paths = raw_data[raw_data['variable'] == 't2m'][['variable', 'data_path']]
t2m_raw_paths.head()
```

```python
# Filter by a specific dataset, e.g., 'derived-era5-single-levels-daily-statistics', and get the paths
derived_era5_raw_paths = raw_data[raw_data['dataset'] == 'derived-era5-single-levels-daily-statistics'][['variable', 'data_path']]
derived_era5_raw_paths.head()
```
